```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Creating the signal

t = np.linspace(0, 1, 300, endpoint=False)
x = np.sin(2 * (np.pi) * 10 * t)
y = 0.5 * (np.sin(2 * (np.pi) * 20 * t))
w = 0.2 * (np.sin(2 * (np.pi) * 50 * t))
z = x + y + w
# for i in range(300):
#     print(f"t = {t[i].round(4)} gives x = {x[i].round(4)}")

# Plotting x, y and w against time t
plt.figure(figsize=(20, 6))
plt.plot(t, x, label="x")
plt.plot(t, y, label="y")
plt.plot(t, w, label="w")

plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title("Plot of x, y, and z against Time")
plt.legend()
plt.grid()

plt.show()
```

```python
# Plotting z(z=x+y+w) agains time t
w_amp_max = np.max(z)
w_amp_min = np.min(z)
# print(f"Max value of w={w_amp_max} & Min value of w={w_amp_min}")
plt.figure(figsize=(20, 6))
plt.plot(t, z, label="z=x+y+z")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title("Plot of the signal z(z=x+y+w) against time T")
plt.legend()
plt.grid()
plt.ylim(w_amp_min, w_amp_max)
# Annotate max and min values on the plot
plt.annotate(
    f"Max: {w_amp_max:.2f}",
    xy=(0, w_amp_max),
    xytext=(0.1, w_amp_max + 0.1),
    arrowprops=dict(facecolor="cyan", shrink=0.05),
)
plt.annotate(
    f"Min: {w_amp_min:.2f}",
    xy=(0, w_amp_min),
    xytext=(0.1, w_amp_min - 0.1),
    arrowprops=dict(facecolor="pink", shrink=0.05),
)
plt.show()
```
