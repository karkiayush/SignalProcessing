```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
```

### We are defining x(t) = cos(2pi.f.t) & plotting from t=(-1,1)

```python
min_t = -1
max_t = 1
count = 1500
t = np.linspace(min_t, max_t, count)
f = 2  # frequency is set as 2 hz

xt = np.cos(2 * np.pi * f * t)
```

```python
# Plotting the signal
plt.figure(figsize=(16, 6))
plt.plot(t, xt, color="red", linewidth=2,label="x(t)=cos(2*pi*f*t)")
plt.legend(loc="upper right",fontsize=14)
plt.xlabel("time in sec", fontsize=15, fontfamily="Monaco")
plt.ylabel("x(t)", fontsize=15, fontfamily="Monaco")
plt.title("Plot of cos(2*pi*f*t)", fontsize=17, fontfamily="Monaco")
plt.grid()
plt.show()
```

```python
# Defining sampling period and sampling frequency
# According to the nyquist theorem: Fs>=2Fm

f_s = 5
t_s = 1 / f_s

# Creating the pulse train for the sampling
pulse_train = np.arange(min_t, max_t, t_s)

# *************arange takes seperation(d) value and linspace takes count(c)*******************
# # arange(x1,x2,sp) returns the value from x1 to x2 at the seperation of sp eg: val1=x1, val2= x1+sp  and so on......
# print(f"\n{np.arange(10,140,4)}")

# # linspace(10,110,10) starts from 10 and it want 10 linearly space element from 10 to 110 and value lie as : 10<=n<=110
# print(f"\n{np.linspace(10,110,10)}")

amplitude = np.ones(len(pulse_train))
plt.figure(figsize=(18,6))
plt.stem(pulse_train, amplitude)
plt.xlabel(
    f"Pulse train ranging from {max_t} to {min_t}", fontsize=15, fontfamily="Monaco"
)
plt.ylabel(f"Amplitude", fontsize=15, fontfamily="Monaco")
plt.title(
    f"Fig: Sampled signal", fontsize=18, fontfamily="Monaco"
)
plt.grid()
plt.show()
```
