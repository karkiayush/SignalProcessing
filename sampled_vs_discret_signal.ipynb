```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Function to generate a discrete signal
def generate_discrete_signal():
    n = np.arange(0, 10)  # Time indices for the discrete signal
    x_discrete = np.array(
        [1, 2, 0, -1, 3, 2, 1, 0, -1, -2]
    )  # Values of the discrete signal
    return n, x_discrete
```

```python
# Function to generate a sampled signal
def generate_sampled_signal():
    t_continuous = np.linspace(0, 10, 1000)  # Time indices for the continuous signal
    x_continuous = np.sin(t_continuous)  # Example continuous signal (sine wave)

    # Sample the continuous signal at discrete points
    # Frequency: no of cycles per second
    sampling_rate = 100  # Number of samples per unit time
    time_period = 1 / sampling_rate
    t_sampled = np.arange(0, 10, time_period)  # Time indices for the sampled signal
    x_sampled = np.sin(t_sampled)  # Sampled values

    return t_continuous, x_continuous, t_sampled, x_sampled
```

```python
# Generate and plot the discrete signal
n, x_discrete = generate_discrete_signal()
plt.stem(n, x_discrete, linefmt="b-", markerfmt="bo", basefmt="r-")
plt.title("Discrete Signal", fontsize=15, fontfamily="Monaco")
plt.xlabel("n", fontsize=13, fontfamily="Monaco")
plt.ylabel("Amplitude", fontsize=13, fontfamily="Monaco")
plt.grid(True)
plt.show()
```

```python
# Generate and plot the sampled signal
t_continuous, x_continuous, t_sampled, x_sampled = generate_sampled_signal()
plt.plot(t_continuous, x_continuous, label="Continuous Signal")
plt.stem(
    t_sampled,
    x_sampled,
    linefmt="g-",
    markerfmt="go",
    basefmt="r-",
    label="Sampled Signal",
)
plt.title("Sampled Signal", fontsize=15, fontfamily="Monaco")
plt.xlabel("Time", fontsize=13, fontfamily="Monaco")
plt.ylabel("Amplitude", fontsize=13, fontfamily="Monaco")
plt.legend()
plt.grid(True)
plt.show()
```
