# Impulse Function plot
##### An impulse function, often denoted as (t) in continuous time or [n] in discrete time, is a mathematical function that is used in signal processing, control systems, and various branches of physics and engineering. It is also known as the Dirac delta function. 



```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
t=[-2,-1,0,1,2]
d=[0,0,1,0,0]

plt.figure(figsize=(20,8))
plt.stem(t,d)
plt.title('Discrete Time impulse function',fontsize=15,fontfamily="Monaco")
plt.xlabel("time(t)",fontsize=15,fontfamily="Monaco")
plt.ylabel("d[t]",fontsize=15,fontfamily="Monaco")
plt.grid(True)
plt.show()
```
