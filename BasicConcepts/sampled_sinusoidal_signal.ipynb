# Plotting Sampled Sine and Cosine Signal


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Defining the independent variable and dependent variable
n = np.arange(0, 1, 0.01)
xn = np.sin(2 * np.pi * 3 * n)
```

```python
# Plotting the discrete signal(sampled signal) that's why we are using the stem function
plt.figure(figsize=(20, 8)) #figure dimension setup

# X and Y axis label and title of wave
plt.xlabel("--------> n", fontsize=15, fontfamily="Monaco")
plt.ylabel("sin(angle)", fontsize=15, fontfamily="Monaco")
plt.title("Sample sine wave", fontsize=15, fontfamily="Monaco")

# For the sampled signal & Discrete signal 
plt.stem(n, xn, label="Sin(angle)")

plt.legend(loc="upper right", fontsize=15)
plt.grid(True)

plt.show()
```
