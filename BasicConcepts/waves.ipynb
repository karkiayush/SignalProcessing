# Plotting sine & cosine wave


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Adjust the range to cover 0 to 180 degrees
# Defining the independent variable
t = np.arange(0, 361, 1)  # Assuming a step size of 1 degree
# Defining the dependent variable
angle_val = np.sin(np.radians(t))  # Convert degrees to radians before applying sine

# Plot the sine function
plt.figure(figsize=(20, 10))
plt.plot(t, angle_val,label="sin(angle)",color="red",linewidth=2,linestyle="-")
plt.title(
    "Sine Function from 0 to 180 degrees",
    fontsize=16,
    fontweight="light",
    fontfamily="Monaco",
)
plt.xlabel("Angle (degrees)", fontsize=13, fontweight="bold", fontfamily="Monaco")
plt.ylabel("sin(angle)", fontsize=13, fontweight="bold", fontfamily="Monaco")

plt.axvline(0,color="black",linestyle="--")
plt.axhline(0, color="black", linestyle="--")
plt.xticks(np.arange(0, 361, 45))
plt.grid(True)
plt.legend(loc="upper right",fontsize=13)
plt.show()
```

```python
# Creating the independent variable
t = np.arange(-180, 361, 1)
cos_angle_value = np.cos(np.radians(t))

# Plot with enhanced features
plt.figure(figsize=(20, 6))
plt.plot(
    t, cos_angle_value, label="cos(angle)", color="blue", linestyle="-", linewidth=2
)

# Add grid and set grid style
plt.grid(True, linestyle="--", alpha=0.7)

# Add title and labels
plt.title(
    "Cosine Function for Angle Values from -180 to 360 degrees",
    fontsize=16,
    fontweight="bold",
    fontfamily="Monaco",
)
plt.xlabel("Angle (degrees)", fontsize=14, fontfamily="Monaco")
plt.ylabel("cos(angle)", fontsize=14, fontfamily="Monaco")

# Add legend
plt.legend(loc="upper right", fontsize=13)

# Add Vertical & horizontal line at y=0 for clarity
plt.axvline(0,color="black",linestyle="--")
plt.axhline(0, color="black", linestyle="--")

# Customize tick parameters
plt.xticks(np.arange(-180, 361, 45))
plt.yticks(np.arange(-1, 1.1, 0.2))

# Add a background grid for better readability
plt.grid(True, linestyle="--", alpha=0.5)

# Show the plot
plt.show()
```
